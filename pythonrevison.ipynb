```python
#Second Largest Number in a List
# Second Largest Number in a List

numbers = [10, 5, 20, 8]
flargest = numbers[0]
selargest = numbers[0]
for i in numbers:
    if i > flargest:
        selargest = flargest
        flargest = i
    elif i > selargest and i != flargest:
        selargest = i

print("Largest =", flargest)
print("Second Largest =", selargest)
```
